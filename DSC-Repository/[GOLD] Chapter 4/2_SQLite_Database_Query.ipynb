## Read SQLite Table and create the dataframe

As we know, to read data from SQLite, we usually create connection, create cursor then read the data, this will return the result data in list, to create dataframe from it, we can use `cursor.description`

**To Do:** Select * from taxi_data table by limit 10, get the column, and create dataframe from it

```python
conn = sqlite3.connect('database/taxi.db')
```

```python
import pandas as pd
# Select 10 baris dari table 'taxis' dan ambil column nya

cursor = conn.execute('SELECT * FROM taxis LIMIT 10')
columns = [col[0] for col in cursor.description]

for col in cursor.description:
    print(col[0])

# buat dataframe dari data yang dipilih
data = cursor.fetchall()
taxi_df = pd.DataFrame(data, columns=columns)

# Close the connection to the database
conn.close()
```

**To Do:** Create Python Function that automatically read query and output dataframe

```python
def cekdataframe(db_file, query):
    #membuat koneksi sqlite3
    conn = sqlite3.connect(db_file)

    #query mengambil 10 data
    cursor = conn.execute(query)
    columns = [col[0] for col in cursor.description]

    #tampung ke dalam dataframe
    data = cursor.fetchall()
    df = pd.DataFrame(data, columns=columns)

    # Tutup koneksi db
    conn.close()

    return df
```

```python
# Panggil fungsi
df = cekdataframe('SELECT * FROM taxis LIMIT 10', 'database/taxis.db')
df
```

```python
import sqlite3
import pandas as pd

# buat koneksi
conn = sqlite3.connect('database/taxis.db')

# define database
query = '''SELECT passengers, distance, fare, tip, tolls FROM taxis 
WHERE pickup_borough = 'Manhattan' 
AND tip > (
    SELECT AVG(tip)
    FROM taxis
    WHERE pickup_borough = 'Queens'
)
'''

# gunakan fungsi read_sql
taxi_df = pd.read_sql(query, conn)

# tutup koneksi
conn.close()

# cek hasil dataframe
taxi_df
```

```python
# Load data dari sql
conn = sqlite3.connect('database/taxis.db')
taxi_df = pd.read_sql_query("SELECT * FROM taxis", conn)

# filter bedasarkan warna taxi
yellow_taxi_df = taxi_df[taxi_df['color'] == 'yellow']

# grouping data berdasarkan jarak max, rata2 rip, dan rata2 penumpang
result_df = yellow_taxi_df.groupby('dropoff_borough').agg(
    {'distance': 'max', 
     'fare': 'mean', 
     'passengers': 'mean'
     })

conn.close()
# view dataframe
result_df
```

```python
conn = sqlite3.connect('database/taxis.db')
# Ambil data dari database
query = ''' SELECT * FROM taxis '''

# fungsi read_sql
taxis_df = pd.read_sql_query(query, conn)

# Hitung rata-rata tip dari semua transaksi
avg_tip = taxis_df['tip'].mean()

# Filter hanya transaksi dengan total tip di atas rata-rata
above_avg_tip_df = taxis_df[taxis_df['total'] > avg_tip]

# Filter hanya transaksi dengan jarak rata-rata di atas 10 km
above_avg_distance_df = above_avg_tip_df[above_avg_tip_df['distance'] > 10]

# Hitung rata-rata jarak dan tarif untuk setiap pickup_zone
grouped_df = above_avg_distance_df.groupby('pickup_zone').agg({'distance': 'mean', 'fare': 'mean'})

# Urutkan berdasarkan nilai tarif rata-rata tertinggi
result_df = grouped_df.sort_values('fare', ascending=False)

conn.close()

# Tampilkan hasil dalam bentuk dataframe
result_df
```

```python
# define database
conn = sqlite3.connect('database/taxis.db')

query = '''SELECT pickup_zone, AVG(distance) AS avg_distance, AVG(fare) AS avg_fare 
         FROM taxis 
         WHERE total > (SELECT AVG(tip) FROM taxis) 
         AND distance > 10 
         GROUP BY pickup_zone
         ORDER BY fare DESC'''

# gunakan fungsi read_sql
taxi_df = pd.read_sql(query, conn)

# tutup koneksi
conn.close()

# cek hasil dataframe
taxi_df
```
