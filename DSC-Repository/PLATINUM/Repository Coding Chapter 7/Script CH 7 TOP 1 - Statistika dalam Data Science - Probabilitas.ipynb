#### Events and Sample Spaces

Asumsikan kita memiliki koin yang sama-sama memiliki kemungkinan muncul kepala/head (H) atau ekor/tail (T).


Dalam kasus seperti ini, di mana dua hasil memiliki kemungkinan yang sama, kita dapat menggunakan teori probabilitas untuk menyatakan kemungkinan **peristiwa/event** tertentu dengan membandingkannya dengan **ruang sampel/sample space** (kumpulan semua hasil yang mungkin dapat dilambangkan sebagai $\Omega$):

$$ P(\text{event}) = \frac{\text{# of outcomes of event}}{\text{# of outcomes in }\Omega} $$

Jika kita hanya melempar koin satu kali, maka hanya ada dua kemungkinan hasil di ruang sampel $\Omega$: itu akan menjadi H atau T (menggunakan notasi himpunan, kita dapat menulis ini sebagai $\Omega$ = {H , T}).

Karena: $$ P(H) = \frac{1}{2} = 0.5 $$

Sama dengan: $$ P(T) = \frac{1}{2} = 0.5 $$

Contoh lain, asumsi terdapat satu kartu dari setumpuk 52 kartu. Dalam hal ini, jumlah hasil yang mungkin dalam ruang sampel (sample space) $\Omega$ adalah 52.

Hanya ada satu kartu as sekop (ace of spades) di deck, sehingga peluang terambilnya adalah: $$ P(\text{ace of spades}) = \frac{1}{52} \approx 0.019 $$

Sebaliknya ada empat Kartu As, sehingga peluang terambilnya Kartu As adalah: $$ P(\text{ace}) = \frac{4}{52} \approx 0.077 $$

Beberapa contoh tambahan:
$$ P(\text{spade}) = \frac{13}{52} = 0.25 $$
$$ P(\text{ace OR spade}) = \frac{16}{52} \approx 0.307 $$
$$ P(\text{card}) = \frac{52}{52} = 1 $$
$$ P(\text{turnip}) = \frac{0}{52} = 0 $$

#### Multiple Independent Observations

Mari kembali ke lempar koin untuk menggambarkan situasi di mana kita memiliki peristiwa yang terdiri dari Multiple Independent Observations. Misalnya, peluang munculnya dua kepala (Head/H) secara berurutan adalah: $$ P(\text{HH}) = \frac{1}{4} = 0.25 $$ ...karena ada satu kejadian (event) HH dalam himpunan sampel empat kemungkinan kejadian (event) ($\Omega$ = {HH, HT, TH, TT}).

Demikian pula, peluang munculnya *tiga* kepala (head) secara berurutan adalah: $$ P(\text{HHH}) = \frac{1}{8} = 0,125 $$ ...karena ada satu kejadian HHH dalam himpunan sampel dari delapan kemungkinan kejadian ($\Omega$ = {HHH, HHT, HTH, THH, HTT, THT, TTH, TTT}).

Sebagai contoh terakhir, peluang terlempar tepat dua kepala dalam tiga kali lemparan adalah $ P = \frac{3}{8} = 0,375 $ sedangkan peluang terlempar setidaknya dua kepala dalam tiga kali lemparan adalah $ P = \frac{4} {8} = 0,5 $.

#### Combining Probabilities

Untuk menggabungkan probabilitas, kita dapat mengalikannya. Jadi peluang munculnya lima kepala berurutan, misalnya, adalah hasil kali dari peluang yang telah kita hitung:$$ P(\text{HHHHH}) = P(\text{HH}) \times P(\text{HHH}) = \frac{1}{4} \times \frac{1}{8} = \frac{1}{32} \approx 0.031 $$

#### Combinatorics

*Combinatorics* adalah bidang matematika yang dikhususkan untuk menghitung yang dapat membantu mempelajari probabilitas. Kita dapat menggunakan **faktorial/factorials** (e.g., $4! = 4 \times 3 \times 2 \times 1 = 24$), yang menonjol dalam Combinatorics adalah untuk menghitung probabilitas daripada susah payah menentukan semua anggota ruang sampel $\Omega$ dan menghitung subset dalam $\Omega$.

Lebih khusus lagi, kita dapat menghitung jumlah hasil dari suatu kejadian menggunakan persamaan "jumlah kombinasi/number of combinations": $$ {n \choose k} = \frac{n!}{k!(n - k)!} $$

The left-hand side of the equation is read "$n$ choose $k$" and is most quickly understood via an example: If we have three coin flips, $n = 3$, and if we're interested in the number of ways to get two head flips (or two tail flips, for that matter), $k = 2$. We would read this as "3 choose 2" and calculate it as:
$$ {n \choose k} = {3 \choose 2} = \frac{3!}{2!(3 - 2)!} = \frac{3!}{(2!)(1!)} = \frac{3 \times 2 \times 1}{(2 \times 1)(1)} = \frac{6}{(2)(1)} = \frac{6}{2} = 3 $$

Sisi kiri persamaan dibaca "$n$ pilih $k$" dan paling cepat dipahami melalui sebuah contoh: Jika kita memiliki tiga pelemparan koin, $n = 3$, dan jika kita tertarik dengan angkanya cara untuk mendapatkan dua flips kepala (atau dua flips ekor, dalam hal ini), $k = 2$. Kita dapat membaca ini sebagai "3 pilih 2" dan menghitungnya sebagai:
$$ {n \choose k} = {3 \choose 2} = \frac{3!}{2!(3 - 2)!} = \frac{3!}{(2!)(1!)} = \frac{3 \times 2 \times 1}{(2 \times 1)(1)} = \frac{6}{(2)(1)} = \frac{6}{2} = 3 $$

Ini memberikan kita pembilang (numerator) untuk persamaan peluang-kejadian (event-probability) berdasarkan kondisi di atas: $$ P(\text{event}) = \frac{\text{# hasil peristiwa}}{\text{# hasil dalam }\Omega} $$

Dalam kasus pelemparan koin (atau proses biner apa pun dengan kemungkinan hasil yang sama), denominator dapat dihitung dengan $2^n$ (di mana $n$ adalah jumlah pelemparan koin), jadi: $$ \frac{\text{# of outcomes of event}}{\text{# of outcomes in }\Omega} = \frac{3}{2^n} = \frac{3}{2^3} = \frac{3}{8} = 0.375 $$

#### The Law of Large Numbers

Sementara banyaknya lemparan koin harus di atas 50% dari suatu waktu, seperti yang telah kita lihat di atas, dengan ukuran sampel yang kecil, ada kemungkinan non-trivial bahwa dalam eksperimen tertentu kita bisa membalik head pada semua lemparan. Misalnya, kita telah menghitung bahwa ada peluang sebesar 3,1% yang akan kita dapatkan pada setiap lemparan dalam percobaan lima lemparan kecil.

**Law of large numbers** menyatakan bahwa semakin banyak eksperimen yang kita jalankan, semakin dekat kita dengan probabilitas yang diharapkan.

Mari kita mencoba mempraktekannya. Pertama panggil library berikut:


*   NumPy untuk mengoperasikan vektor
*   Scipy untuk mengoperasikan operasi statistika
*   Matplotlib dan Seaborn untuk visualisasi 



```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

Untuk memulai, kita akan membuat vektor yang meningkatkan $n$jumlah lemparan koin secara eksponensial per eksperimen:

```python
ns = np.array([2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096]) 
```

Jalankan "np.random.seed" untuk menghasilkan satu set data baru

```python
np.random.seed(42)
```

Kita akan membahas metode `binomial()` secara lebih rinci bagian selanjutnya. Untuk saat ini, cukup untuk memikirkan dua argumennya sebagai *number of coin flips in experiment* dan *probability of heads*, sementara itu mengembalikan jumlah lemparan yang menghasilkan kepala dalam percobaan.

```python
np.random.binomial(1, 0.5)
```

```python
heads_count = [np.random.binomial(n, 0.5) for n in ns]
heads_count
```

```python
proportion_heads = heads_count/ns
proportion_heads
```

```python
fig, ax = plt.subplots()
plt.xlabel('Number of coin flips in experiment')
plt.ylabel('Proportion of flips that are heads')
plt.axhline(0.5, color='orange')
_ = ax.scatter(ns, proportion_heads)
```

Penting pada saat ini untuk mengatasi **gambler's fallacy**. Ini adalah kesalahpahaman umum bahwa hukum bilangan besar menyatakan bahwa jika, katakanlah, lima kepala telah dibalik berturut-turut, maka kemungkinan ekor lebih tinggi pada flip keenam. Faktanya, teori probabilitas menyatakan bahwa setiap lemparan koin benar-benar independen dari yang lainnya. Dengan demikian, setiap lemparan koin yang adil memiliki peluang 50% untuk menjadi kepala, tidak peduli apa yang terjadi pada lemparan sebelumnya.

#### Statistics

Bidang statistik menerapkan teori probabilitas untuk membuat kesimpulan dengan tingkat kepercayaan yang dapat diukur. Misalnya, katakanlah kita menjalankan eksperimen lempar lima koin 1000 kali dengan koin yang adil:

```python
n_experiments = 1000
heads_count = np.random.binomial(5, 0.5, n_experiments)
```

Kita dapat memplot hasil eksperimen untuk membuat **distribusi probabilitas**:

```python
heads, event_count = np.unique(heads_count, return_counts=True)
```

```python
heads
```

```python
event_count
```

```python
event_proba = event_count/n_experiments
event_proba
```

```python
plt.bar(heads, event_proba, color='mediumpurple')
plt.xlabel('Heads flips (out of 5 tosses)')
_ = plt.ylabel('Event probability')
```
