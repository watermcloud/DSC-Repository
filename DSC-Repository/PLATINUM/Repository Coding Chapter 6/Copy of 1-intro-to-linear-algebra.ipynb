### $L^2$ Norm

```python
# Untuk menghitung L2 Norm secara manual menggunakan Python bisa seperti berikut
# Asumsi kita punya vector [25, 2, 5]
# Setelah itu kita hitung berdasarkan Equation, yakni:
# 1. Pangkatkan setiap nilai dari Vector
# 2. Hasil pemangkatan dijumlahkan
# 3. Hasil dari penjumlahan kita ambil akarnya
# Teknisnya sebagai berikut:

(25**2 + 2**2 + 5**2)**(1/2)
```

### Basic Matrices Arithmetic for Hadamard Product

Hadamard Product adalah Jika ada 2 Matrix dengan size yang sama melakukan operasi Aritmatika

Notasi dari operasi ini adalah $A \odot X$

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

import numpy as np
```

```python
# Definisikan sebuah Matrix dengan nested brackets (dobel kurung siku) menggunakan 
# NumPy ke dalam suatu variabel

X = np.array([[25, 2], [5, 26], [3, 7]])
X
```

```python
# Definisikan sebuah Matrix lagi
# Kita beri nama variabel matrix baru ini dengan nama "A"

A = X+2
A
```

```python
# Menambahkan nilai dari Matrix dengan Matrix lainnya
# Hasil penambahan kita masukkan ke variabel "A_X"

A_X = A + X
A_X
```

```python
# Mengurangkan nilai dari Matrix dengan Matrix lainnya
# Hasil pengurangan kita masukkan ke variabel "A_X"

A_X = A - X
A_X
```

```python
# Mengalikan nilai dari Matrix dengan Matrix lainnya
# Hasil perkalian kita masukkan ke variabel "A_X"

A_X = A * X
A_X
```

```python
# Membagikan nilai dari Matrix dengan Matrix lainnya
# Hasil pembagian kita masukkan ke variabel "A_X"

A_X = A / X
A_X
```

### Substitution Demo Example

Sebelumnya kita sudah memberi contoh **Substitution** seperti berikut, kita namakan y1: 
$$ y = 3x $$
$$ -5x + 2y = 2 $$

Selanjutnya kita bisa mengubah equation sebelumnya menjadi seperti ini, dan kita namakan y2: 
$$ 2y = 2 + 5x $$
$$ y = \frac{2 + 5x}{2} = 1 + \frac{5x}{2} $$

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

# Selanjutnya juga kita import Matplotlib untuk visualisasi

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate matrix untuk yang kita masukan ke variabel x

x = np.linspace(-10, 10, 1000) # start, finish, n points
```

```python
# Kita operasikan equation "y=3x" menjadi seperti berikut:

y1 = 3 * x
```

```python
# Kita operasikan equation "1+(5x/2)" menjadi seperti berikut

y2 = 1 + (5*x)/2
```

```python
# Setelah itu kita visualisasikan hasil perhitungan menjaid seperti ini

fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 3])
ax.set_ylim([0, 8])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=2, color='purple', linestyle='--')
_ = plt.axhline(y=6, color='purple', linestyle='--')
```

### Elimination Demo Example

Sebelumnya kita sudah memberi equation untuk **Elimination** seperti berikut: 
$$ 2x - 3y = 15 $$
$$ 4x + 10y = 14 $$

Equations pertama dapat kita lakukan elimination dan kita namakan y1: 
$$ -3y = 15 - 2x $$
$$ y = \frac{15 - 2x}{-3} = -5 + \frac{2x}{3} $$

Selanjutnya equations kedua dapat kita lakukan elimination dan kita namakan y2: 
$$ 4x + 10y = 14 $$
$$ 2x + 5y = 7 $$
$$ 5y = 7 - 2x $$
$$ y = \frac{7 - 2x}{5} $$

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

# Selanjutnya juga kita import Matplotlib untuk visualisasi

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate matrix untuk yang kita masukan ke variabel x

x = np.linspace(-10, 10, 1000) # start, finish, n points
```

```python
# Kita operasikan equation "-5+(2x/3)" menjadi seperti berikut:

y1 = -5 + (2*x)/3
```

```python
# Kita operasikan equation "(7-2x/5)" menjadi seperti berikut:

y2 = (7-2*x)/5
```

```python
# Setelah itu kita visualisasikan hasil perhitungan menjaid seperti ini

fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

# Add x and y axes: 
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-2, 10])
ax.set_ylim([-6, 4])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=6, color='purple', linestyle='--')
_ = plt.axhline(y=-1, color='purple', linestyle='--')
```

### Matrix Inversion

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

import numpy as np
```

```python
# Kita definisikan Matrix dengan menambah kurung siku pada NumPy
# Seperti berikut: 

X = np.array([[4, 2], [-5, -3]])
X
```

```python
# Lakukan Matrix Inversion

Xinv = np.linalg.inv(X)
Xinv
```

```python
# Definisikan vector ke dalam nilai y
# Seperti berikut: 

y = np.array([4, -7])
y
```

```python
# Jalankan Matrix Multiplication pada hasil Matrix Inversion (Xinv) dan variabel y 

w = np.dot(Xinv, y)
w
```

Show that $y = Xw$: 

```python
# Cek apakah y sama dengan Xw
np.dot(X, w)
```
