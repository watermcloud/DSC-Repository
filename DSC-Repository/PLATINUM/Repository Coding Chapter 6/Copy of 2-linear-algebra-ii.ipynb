### Eigenvectors and Eigenvalues

Seperti dari ilustrasi yang sudah kita jabarkan, Eigenvector adalah special vector $v$ yang terbentuk dari hasil transformasi Matrix yang mana special vector $Av$ memiliki arah yang sama dengan $v$.

Eigenvalue adalah sebuah nilai scalar (disimbolkan $\lambda$) dari Eigenvector $v$.

Jika ditulis ke dalam sebuah equation akan menjadi seperti ini:

$Av = \lambda v$

Sebagai contoh: 

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

import numpy as np
```

```python
# Definisikan sebuah Matrix menggunakan NumPy ke dalam suatu variabel
# Matrix ini memiliki nilai [1, 2], [3, 4.]

A = np.array([[1, 2], [3, 4]])
A
```

```python
# Untuk menghitung Eigen, kita akan menggunakan fungsi "np.linalg.eig"
# Fungsi di atas akan mengeluarkan sebuah output sebagai berikut:
# 1. a vector of eigenvalues (lambdas)
# 2. a matrix of eigenvectors (V)

lambdas, V = np.linalg.eig(A) 
```

```python
# Eigenvector yang terdeteksi dari Matrix yang diproses
# Sebagai berikut:

V
```

```python
# Eigenvalue dari Eigenvector
# Sebagai berikut:

lambdas
```

Konfirmasi bahwa $Av = \lambda v$ untuk Eigenvector yang pertama: 

```python
# Pertama kita ambil nilai dari Eigenvector pertama
# Sebagai berikut:

v = V[:,0] 
v
```

```python
# Kedua, kita ambil lambda (Eigenvalue) dari Eigenvector pertama
# Sebagai berikut:

lambduh = lambdas[0] # note that "lambda" is reserved term in Python
lambduh
```

```python
# Ketiga, kita lakukan Matrix Multiplication untuk mendapatkan Av
# Sebagai berikut:

Av = np.dot(A, v)
Av
```

```python
# Selanjutnya, kita lakukan Matrix Multiplication juga untuk mendapatkan "Lambda V"
# Sebagai berikut:

lambduh * v
```

```python
# Hasilnya terkofirmasi bahwa "Av" samadengan "Lambda V"
# "Av" => array([ 0.30697009, -0.21062466])
# "Lambda V" =>  array([ 0.30697009, -0.21062466])
```

Selanjutnya kita akan memvisualissikan perhitungan di atas.

```python
# Import library Matplotlib dan buat sebuah fungsi untuk visualisasi vector

import matplotlib.pyplot as plt

def plot_vectors(vectors, colors):
    """
    Plot one or more vectors in a 2D plane, specifying a color for each. 

    Arguments
    ---------
    vectors: list of lists or of arrays
        Coordinates of the vectors to plot. For example, [[1, 3], [2, 2]] 
        contains two vectors to plot, [1, 3] and [2, 2].
    colors: list
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
        
    Example
    -------
    plot_vectors([[1, 3], [2, 2]], ['red', 'blue'])
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    """
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

```python
# Selanjutnya kita visualisasikan Eigenvector dan Eigenvalue dari Matrix/Vector yang pertama
# Warna "blue" merujuk ke Vector yang pertama
# Warna "lightblue" merujuk ke Eigenvector dari Vector yang pertama
# Yang mana Eigenvector dari Vector yang pertama memiliki nilai Eigenvalue -0.37

plot_vectors([Av, v], ['blue', 'lightblue'])
plt.xlim(-1, 2)
_ = plt.ylim(-1, 2)
```

Selanjutnya, konfirmasi bahwa $Av = \lambda v$ untuk Eigenvector yang kedua: 

```python
# Pertama kita ambil nilai dari Eigenvector kedua
# Sebagai berikut:

v2 = V[:,1]
v2
```

```python
# Selanjutnya, kita lakukan Matrix Multiplication juga untuk mendapatkan "Lambda V"
# Sebagai berikut:

lambda2 = lambdas[1]
lambda2
```

```python
# Ketiga, kita lakukan Matrix Multiplication untuk mendapatkan Av
# Sebagai berikut:

Av2 = np.dot(A, v2)
Av2
```

```python
# Selanjutnya, kita lakukan Matrix Multiplication juga untuk mendapatkan "Lambda V"
# Sebagai berikut:

lambda2 * v2
```

```python
# Selanjutnya kita visualisasikan Eigenvector dan Eigenvalue dari Matrix/Vector yang pertama dan yang kedua
# Warna "blue" merujuk ke Vector yang pertama
# Warna "lightblue" merujuk ke Eigenvector dari Vector yang pertama
# Yang mana Eigenvector dari Vector yang pertama memiliki nilai Eigenvalue -0.37

# Warna "green" merujuk ke Vector yang kedua
# Warna "lightgreen" merujuk ke Eigenvector dari Vector yang kedua
# Yang mana Eigenvector dari Vector yang kedua memiliki nilai Eigenvalue 5.37

plot_vectors([Av, v, Av2, v2], 
            ['blue', 'lightblue', 'green', 'lightgreen'])
plt.xlim(-1, 4)
_ = plt.ylim(-3, 2)
```

### Eigendecomposition

Equation **eigendecomposition** sebagai berikut: 

$A = V \Lambda V^{-1}$

Dimana: 

* $V$ adalah gabungan dari semua Eigenvectors dari $A$
* $\Lambda$ (upper-case $\lambda$) atau Lambda adalah diagonal matrix diag($\lambda$).

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

import numpy as np
```

```python
# Definisikan sebuah Matrix menggunakan NumPy ke dalam suatu variabel
# Matrix ini memiliki nilai [4, 2], [-5, -3]

A = np.array([[4, 2], [-5, -3]]) 
A
```

```python
# Cari nilai Eigenvector dan Eigenvalue nya

lambdas, V = np.linalg.eig(A)
```

```python
# Berikut nilai Eigenvector nya

V
```

```python
# Cari nilai Inverse dari V

Vinv = np.linalg.inv(V)
Vinv
```

```python
# Cari Diagonal Matrix

Lambda = np.diag(lambdas)
Lambda
```

Cek apakah nilai yang kita dapatkan di atas sesuai persamaan berikut 
$A = V \Lambda V^{-1}$: 

```python
# Setelah menjalankan perintan berikut, kita mendapatkan nilai:
# [ 4.,  2.], [-5., -3.]
# Nilai di atas sama dengan nilai dari Matrix A yang kita definisikan, yaitu:
# [ 4,  2], [-5, -3]
# Ketika nilainya sama pertanda bahwa kita sudah berhasil melakukan Eigendecomposition

np.dot(V, np.dot(Lambda, Vinv))
```

Eigendecomposition tidak bisa diterapkan pada semua Matrix. Dan pada kasus tertentu Eigendecomposition melibatkan angka yang lebih kompleks.

Dalam beberapa kasus Machine Learning, kita akan sering menemui Matrix berbentuk real symmetric matrices, yang mana dapat dengan mudah dan efisien ketika akan dilakukan decomposed ke real-only eigenvectors real-only eigenvalues. If $A$ adalah real symmetric matrix, maka

$A = Q \Lambda Q^T$


```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

import numpy as np
```

```python
# Definisikan sebuah Real Symmetric Matrix menggunakan NumPy ke dalam suatu variabel
# Matrix ini memiliki nilai [2, 1], [1, 2]

A = np.array([[2, 1], [1, 2]])
A
```

```python
# Cari nilai Eigenvector dan Eigenvalue nya

lambdas, Q = np.linalg.eig(A)
```

```python
# Berikut nilai Eigenvector nya

lambdas
```

```python
# Berikut nilai Matrix Diagonal nya

Lambda = np.diag(lambdas)
Lambda
```

```python
# Berikut nilai Eigenvector nya

Q
```

Cek apakah nilai yang kita dapatkan di atas sesuai persamaan berikut $A = Q \Lambda Q^T$:

```python
# Setelah menjalankan perintan berikut, kita mendapatkan nilai:
# [2, 1], [1, 2]
# Nilai di atas sama dengan nilai dari Matrix A yang kita definisikan, yaitu:
# [2, 1], [1, 2]
# Ketika nilainya sama pertanda bahwa kita sudah berhasil melakukan Eigendecomposition

np.dot(Q, np.dot(Lambda, Q.T))
```
