## Cleaning Missing Value from the data

### Dropping missing values
```python
# Use dropna to drop the missing values
df.dropna()
```
### Filling Missing Data with constant values
```python
# Use fillna to fill the missing values
fill_value = 0
df.fillna(fill_value)
```

### Filling with method
Fillna function has built in method to fill the missing value, like `{'backfill', 'bfill', 'pad', 'ffill'}`

**To Do:** check `fillna()` help function and discover what are thos method function



````python
# ini manual
fill_value = 0
df.fillna(fill_value)
````

````python
df.ffill()
````

````python
df.backfill()
````

***backfill atau bfill*** akan mengisi nilai-nilai yang hilang dengan nilai observasi berikutnya dalam DataFrame atau Series.

***pad atau ffill*** akan mengisi nilai-nilai yang hilang dengan nilai observasi sebelumnya dalam DataFrame atau Series.

**ffill vs pad** adalah ffill lebih umum dipakai dan merupakan singkatan dari "forward fill". Keduanya bisa dipakai, tergantung pada preferensi kita sebagai programmer.
