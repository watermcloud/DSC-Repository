# Operators

Operators adalah suatu simbol yang digunakan untuk operasi tertentu

Operators dibagi menjadi 4:


*   Arithmetic Operators
*   Assignment Operators
*   Comparisson Operators
*   Logical Operators



**Arithmetic Operators** digunakan ketika ingin mengoperasikan operasi matematika

```
# Pertama kita mendefinsikan variabel dengan nilai tertentu

x = 14
y = 4
```

```
# Penjumlahan
print(x+y) # Output: x + y = 18
```

```
# Pengurangan
print('x - y =', x-y) # Output: x - y = 10
```

```
# Perkalian
print('x * y =', x*y) # Output: x * y = 56
```

```
# Pembagian
print('x / y =', x/y) # Output: x / y = 3.5
```

```
# Perpangkatan
print('x ** y =', x**y) # Output: x ** y = 38416
```

```
a = '4'
b = '5'

print(int(a)*int(b))
```

**Assignment Operators** digunakan ketika ingin menetapkan nilai dalam suatu variable

```
# Perintah "=" digunakan untuk memasukkan nilai dalam suatu variabel

x = 5
```

```
# Selanjutnya "+=", perintah ini untuk menambahkan nilai ke dalam variabel yang sudah ada
x += 5

print(x)
```

```
# Selanjutnya "-=", perintah ini untuk mengurangkan nilai ke dalam variabel yang sudah ada
x -= 1
print(x)
```

```
# Selanjutnya "/=", perintah ini untuk membagi nilai ke dalam variable yang sudah ada
x /= 3
print(x)
```

**Comparisson Operators**

```
a = 9
b = 8
```

```
# ==
# Bernilai sama
print(a == b)
```

```
# >
# Lebih dari
print(a > b)
```

```
# <
# Kurang dari
print(a < b)
```

```
# >=
# Lebih dari sama dengan
print(a >= b)
```

```
# <=
# Kurang dari sama dengan
print(a <= b)
```

**Logical Operators**

```
a = 7
b = 8
```

```
# and
# Jika keduanya benar, maka True
print(a == 8 and b == 8)
```

```
# or
# Jika salah satu benar, maka True
print(a == 7 or b == 7)
```

# Data Structures

Untuk memanajemen suatu kumpulan data, Python memiliki beberapa Data Structures

**Lists**

List adalah suatu tempat penyimpanan data di dalam sebuah kurung siku (square bracket) "[]" yang dipisahkan oleh koma. Data yang simpan dapat berbentuk Integer, Float, dan String.

```
# List dengan nilai kosong
my_list = []
print(my_list)
```

```
# List dengan nilai integer dan string
my_list = [1, "toni", 3]
print(my_list)
```

```
# Tambah data
my_list.append(4)
print(my_list)
```

```
# Hapus data
my_list.pop()
print(my_list)
```

```
# Edit data
my_list[1] = 'tino'
print(my_list)
```

**Tuples**

Sama dengan Lists, perbedaannya dengan Tuples kita tidak dapat mengubah isi datanya ketika sudah didefinsikan. Selain itu Tupples didefinisikan tidak menggunakan kurung siku "()"

```
# Tuples dengan nilai kosong
my_tuples = ()
print(my_tuples)
```

```
# Tuples dengan nilai integer dan string
my_tuples = (1, "toni", 3)
print(my_tuples)
```

**Sets**

Sets adalah tempat penyimpanan dari kumpulan *unique* data.

```
# Mengambil nilai unik dari kumpulan data integer
my_set = {1, 2, 3, 3}
print(my_set)
```

```
# Mengambil nilai unik dari kumpulan data string
my_set = {'sabrina', 'ikhwan', 'ikhwan', 'afif', 'afif', 'afif'}
print(my_set)
```

```
# Set juga dapat dilakukan untuk mendapat nilai 
# unique dari suatu kumpulan list

my_list = ["eka", "akbar", "akbar", "sabrina", "eka", "lina", 
           "ilham", "lina"]

print(set(my_list))
```

**Dictionaries**

Jika tempat penyimpanan sebelumnya hanya dapat menyimpan *value* saja, ketika memakai Dictionaries ini kita juga dapat menyimpan *key* dan *value*.

```
# Dictionaries dengan nilai kosong
my_dict = {}
print(my_dict)
```

```
# Menambahkan data ke Dictionaries
my_dict['name'] = 'Sabrina'
my_dict['age'] = 30
print(my_dict)
```

```
# Mengakses data dari Dictionaries
print(my_dict['name'])
print(my_dict['age'])
```

```
# Mengedit data dari Dictionaries
my_dict['age'] = 25
print(my_dict['age'])
```

```
# Menghapus data dari Dictionaries
del my_dict['age']
print(my_dict)
```

```
# Dictionary dengan keys Integer dan String
# my_dict = {'name': 'John', 'grade': [2, 4, 3]}

data = {'students': 
          [{
            'name': 'Sabrina',
            'age': 20
          },
          {
            'name': 'Bima',
            'age': 25
          },
          {
            'name': 'Yafi',
            'age': 29
          }]
        }
```

```
print(data['students'])
data['students'][0] = {'name': 'Dika', 'age': 20}
print(data['students'])
```

```
print(data)
data['students'] = 'halo halo'
print(data)
```

# If...Else Statement (Percabangan)

If...Else Statement digunakan ketika akan melakukan aksi berbeda dengan kondisi tertentu

```
alamat = 'sleman'

if alamat == 'jogja':
  print('boleh masuk!')
elif alamat == 'sleman':
  print('hmm boleh lah..')
else:
  print('periksa dulu')
```

```
num = -1

if num > 0:
  print("Angka positif")
elif num == 0:
  print("Nol")
else:
  print("Angka negatif")
```

# While Loop Statement (Perulangan)

While Loop Statement digunakan untuk melakukan iterasi kumpulan data selama kondisinya terpenuhi.

```
n = 2

while(n <= 15):
  print(n)
  n += 1
```

# For Loop Statement (Perulangan)

For Loop Statement digunakan untuk melakukan iterasi kumpulan data.

```
names = ["ari", "doni", "dino"]

for name in names:
  print(name)
```

```
sum = 0
numbers = [6, 5, 3, 8, 4, 2]
# Iterasi kumpulan data
for number in numbers:
  print(number)
  sum = sum+number
  print("hasil: ", sum)
```

# Break Statement

Break Statement digunakan untuk menyetop suatu proses perulangan.

```
for val in "binar":
    print(val)
    if val == "a":
        break
```

# Continue Statement

Continue Statement digunakan untuk melakukan skip suatu code.

```
for val in "binar":
    if val == "a":
        continue
    print(val)
```

# Pass Statement

```
sequence = ['a', 'b', 'c']
for val in sequence:
  if val == 'c':
    print("oke")
  else:
    pass
```

# Function

Function atau Fungsi adalah sebuah kumpulan script untuk spesifik task

```
def ages():
  print(25)
  print(26)

def names():
  print("ari")
  print("tino")
```

```
names()
ages()

# def print_lines():
#   print("Cetak 1.")
#   print("Cetak 2.")

# menjalankan fungsi
# print_lines()
```

**Fungsi tanpa Parameter**

```
def add_numbers():
  sum = 1 + 2
  return sum


add_numbers()
```

**Fungsi dengan Parameter**

```
def add_numbers(a, b):
  sum = a + b
  return sum

add_numbers(1, 2)
```

**Fungsi dengan Default Parameter**

```
def add_numbers(a=1, b=2):
  sum = a + b
  return sum
```

```
add_numbers()
```

```
add_numbers(3, 3)
```

# Lambda Function

Lambda: Fungsi yang tidak memiliki nama

# File I/O

File I/O perintah dalam Python yang digunakan ketika berhubungan dengan file. Diantaranya:


1.   Write File
2.   Open File



**Write File**

```
with open("names.txt",'w',encoding = 'utf-8') as f:
   f.write("Ari\n")
   f.write("Dino\n")
   f.write("Dono\n")
```

**Read File**

```
f = open("names.txt",'r',encoding = 'utf-8')
print(f.read())
```
